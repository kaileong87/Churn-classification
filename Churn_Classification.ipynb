we are now ready to run our models, we will try out Decision Tree, Adaptive Boosting, Gradient Boosting, Random Forest and finally Logistic Regression model

```python
from sklearn.model_selection import train_test_split
```

```python
df
```

```python
X = df.drop('Churn', axis=1)
```

```python
X = pd.get_dummies(X, drop_first=True)
```

```python
X.columns
```

```python
y = df['Churn']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=101)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
tree = DecisionTreeClassifier(max_depth=6)
```

```python
tree.fit(X_train, y_train)
```

```python
preds = tree.predict(X_test)
```

```python
from sklearn.metrics import plot_confusion_matrix, confusion_matrix, accuracy_score, classification_report
```

```python
print(classification_report(y_test, preds))
```

```python
plot_confusion_matrix(tree, X_test, y_test)
```

```python
from sklearn.tree import plot_tree
```

```python
plt.figure(figsize=(12,6), dpi=200)
plot_tree(tree, filled=True, feature_names=X.columns);
```

Since we'll be running a few models, let's create a function to automate the whole process

```python
def run_model(model, X_train, X_test, y_train, y_test):
    model.fit(X_train, y_train)
    preds = model.predict(X_test)
    
    print(classification_report(y_test, preds))
    plot_confusion_matrix(model, X_test, y_test)
```

```python
from sklearn.ensemble import AdaBoostClassifier
```

```python
Ada_model = AdaBoostClassifier()
```

```python
param_grid = {'n_estimators': [25, 50, 75, 100, 200, 300],
             'learning_rate': [0.1, 0.2, 0.5, 1]}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid = GridSearchCV(Ada_model, param_grid, cv=5, scoring='balanced_accuracy')
```

```python
run_model(grid, X_train, X_test, y_train, y_test)
```

```python
grid.best_estimator_
```

```python
grid.best_estimator_.feature_importances_
```

```python
feat_imp = pd.DataFrame(data=grid.best_estimator_.feature_importances_, index=X.columns, columns=['importance'])
```

```python
plt.figure(figsize=(12,8), dpi=200)
sns.barplot(data=feat_imp.sort_values('importance'), x=X.columns, y='importance')
plt.xticks(rotation=90);
```

```python
def feat_imp_plot(model):
    feat_imp = pd.DataFrame(data=model.best_estimator_.feature_importances_, index=X.columns, columns=['importance'])
    plt.figure(figsize=(12,8), dpi=200)
    sns.barplot(data=feat_imp.sort_values('importance'), x=X.columns, y='importance')
    plt.xticks(rotation=90);
```

```python
from sklearn.ensemble import GradientBoostingClassifier
```

```python
gradient = GradientBoostingClassifier()
```

```python
param_grid = {'n_estimators': [25, 50, 75, 100, 200, 300],
             'learning_rate': [0.1, 0.2, 0.5, 1],
             'max_depth': [2,3,4,5,6]}
```

```python
gradient_grid = GridSearchCV(gradient, param_grid, cv=5, scoring='balanced_accuracy')
```

```python
run_model(gradient_grid, X_train, X_test, y_train, y_test)
```

```python
feat_imp_plot(gradient_grid)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier()
```

```python
param_grid = {'n_estimators': [25, 50, 75, 100, 200, 300],
             'max_depth': [3,4,5,6,7,8],
             'max_features': ['auto', 'log2'],
             'bootstrap': ['True', 'False'],
             'criterion': ['gini', 'entropy']}
```

```python
rfc_grid = GridSearchCV(rfc, param_grid, cv=5, scoring='balanced_accuracy')
```

```python
run_model(rfc_grid, X_train, X_test, y_train, y_test)
```

```python
feat_imp_plot(rfc_grid)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
log_model = LogisticRegression()
```

```python
param_grid = {'penalty': ['l1', 'l2', 'elasticnet'],
             'C': [0.001, 0.01, 0.1, 0.5, 1, 10, 100],
             'solver': ['saga'],
             'max_iter': [3000],
             'l1_ratio': [0.1, 0.3, 0.5, 0.7, 0.9]}
```

```python
log_grid = GridSearchCV(log_model, param_grid, cv=5, scoring='balanced_accuracy')
```

```python
run_model(log_grid, X_train, X_test, y_train, y_test)
```

```python
log_grid.best_estimator_
```

All 5 models give roughly the same performance. The engineered feature of tenure cohort provides the greatest importance in understanding Churn rate
